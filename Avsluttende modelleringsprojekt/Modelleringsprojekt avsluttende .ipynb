# Avsluttende modelleringsprojekt
## Smittemodellering 

### 1. Hensikt
Her skal vi undersøke en såkalt SIR-modell for smitteutvikling, samt gjennomføre en regresjonsanalyse av ekte smittetall, for så å sammenlikne disse med den koronasituasjonen vi nå befinner oss i.

### 2. Teori

#### Modell 1:
Den første modellen kalles en SIR-smittemodell, og deler populasjonen N inn i tre grupper; S(t) (de som er mottakelige for sykdommen, eller susceptibles), I(t) (de smittede, eller infected) og R(t) (de som har blitt friske fra sykdommen, eller recovered). Modellen undersøker hvordan disse tre gruppene utvikler seg etter t-te dag. For å finne endringene i de forskjellige gruppene mellom to tidssteg dt, bruker har vi noen formler. For endringen i antall mottakelige bruker vi:
$$ S'(t) = -b * S(t) * I(t) $$
Her er b er en parameter som bestemmer spredningsrate, altså hvor mange hver syke person smitter.

Endringen i hvor mange som blir friske ved hvert tidssteg baseres på hvor mange som for øyeblikket er syke, og på bedringsraten f. Dette kan vi skrive slik:
$$ R'(t) = f * I(t) $$
Ettersom de tre gruppene tilsammen dekker hele populasjonen vil $ I'(t)+S'(t)+R'(t)=0$, så lenge N er konstant. Dette kan vi skrive om til $I'(t)=-S(t)-R(t)$, som er hvordan man kommer frem til formelen:
$$ I'(t) = b * S(t) * I(t) - f * I(t) $$ 

#### Modell 2:
Modell 2 likner veldig mye på modell 1, men her har vi gjort om parameterene b og fra fra konstanter, til funksjoner uttrykt ved tiden t.

#### Modell 3:
I modell 3 baserer vi oss ikke lenger på en SIR modell, men bruker allerede eksisterende data til å undersøke mulige videre utviklinger ved hjelp av regresjon. Basert på datapunker vi har fått ved observasjon lager vi et tilpasset polynom som vi kan basere våre tanker om videre utvikling på. 

### 3. Resultater

#### Modell 1:

```python
from pylab import *

#konstanter
N = 124000 #befolkningstall
b = 2e-5 #smittetall
f = 0.2 #friskhetsrate
dt = 1e-2
t_slutt = 10 #dager

#startbetingelser
t0 = 0
I0 = 10 #antall syke ved t=0
S0 = N - I0 #antall mottakelige ved t=0
R0 = 0 #antall friske igjen ved t=0

t = t0
I = I0
S = S0
R = R0

#lister
tlst = [t0] #tid
Ilst = [I0] #smittede
Slst = [S0] #mottakelige
Rlst = [R0] #friske

#eulers metode
for i in arange(0,t_slutt,dt):
    Sd = -b * S * I
    Rd = f * I
    Id = -Sd -Rd
    
    S += Sd * dt
    I += Id * dt
    R += Rd *dt
    t = t + dt
    
    tlst.append(t)
    Slst.append(S)
    Ilst.append(I)
    Rlst.append(R)

#plot
plot(tlst,Ilst,label="Smittede")
plot(tlst,Slst,label="Mottakelig")
plot(tlst,Rlst,label="Friskemeldte")
xlabel("Dager")
ylabel("Individer")
title("Eksempel på SIR-smittemodell")
legend()
grid()
show()
    
```

#### Modell 2:

```python
from pylab import *
    
#funksjoner
def b(t):
    return 2E-5*(t+1)*exp(-0.01*t**2)

def f(t):
    if t < 4:
        return 0.2
    else:
        return 0.4 + 0.5*sin(0.6*pi*t)
    
def eulers(b,f,S0,I0,R0,t_slutt):
    t = t0
    I = I0
    S = S0
    R = R0
    tlst = [t0]
    Ilst = [I0]
    Slst = [S0]
    Rlst = [R0]

    for i in arange(0,t_slutt,dt):
        Sd = -b(t) * S * I
        Rd = f(t) * I
        Id = -Sd -Rd

        S += Sd * dt
        I += Id * dt
        R += Rd *dt
        t = t + dt

        tlst.append(t)
        Slst.append(S)
        Ilst.append(I)
        Rlst.append(R)
    return tlst,Slst,Ilst,Rlst

#konstanter
N = 124000 #befolkningstall
dt = 1e-2
t_slutt = 10 #dager

#startbetingelser
t0 = 0
I0 = 10 #antall syke ved t=0
S0 = N - I0 #antall mottakelige ved t=0
R0 = 0 #antall friske igjen ved t=0


tid,mottakelige,smittede,friske = eulers(b,f,S0,I0,R0,t_slutt)  #henter verdiene fra funksjoner eulers 

#plotting
plot(tid,smittede,label="Smittede")
plot(tid,mottakelige,label="Mottakelige")
plot(tid,friske,label="Friskemeldte")
xlabel("Dager")
ylabel("Individer")
title("SIR-smittemodell med f og b som funksjon av tiden")
legend()
grid()
show()
```

#### Modell 3:

```python
from pylab import *

#laster ned data fra FHI
data = loadtxt("corona-mars-juli-2020.txt",usecols=range(1,3))
smittede = data[:,1]
smittedetot = data[:,0]

smittede = smittede.tolist()
smittedetot = smittedetot.tolist()

smittede.reverse() #reverserer liste for å få verdier i riktig rekkefølge
smittedetot.reverse()

dager = len(smittede)
tid = linspace(0,dager,dager) #lager verdier for x-aksen

#plotter
plot(tid,smittede,label="Nye smittede")
xlabel("Dager fra 01.03.2020")
ylabel("Individer")
title("Antall nye smittede per dag fra 01.03.2020-31.07.2020")
legend()
grid()
show()

plot(tid,smittedetot,label="Totalt antall smittede")
xlabel("Dager fra 01.03.2020")
ylabel("Individer")
title("Totalt antall smittede fra 01.03.2020-31.07.2020")
legend()
grid()
show()
```

Bruker datasettet til å lage en regresjonsanalyse

```python
smittede_reg = smittedetot[:20] #velger de første dagene, før smitten begynner å stagnere for regresjonsanalyse
dager_reg = array(range(len(smittede_reg)))

a,b,c,d,e = polyfit(dager_reg,smittede_reg,4) #finner koeffisienter til tilnæmet polynom

def tilpasning(x):
    return a*x**4 + b*x**3 + c*x**2 + d*x + e #tilpasset polynom

#regresjon med polynom tilpasset observerte data
y = tilpasning(dager_reg)

#plotter regresjon
plot(dager_reg,y,label="Tilpasset kurve")
plot(dager_reg,smittede_reg,linestyle="",marker=".",label="Observerte data")
xlabel("Dager")
ylabel("Smittede individer")
title("Regresjon av data")
grid()
legend()
show()

#utvider tidsrom for å se når vi når 2,2 millioner smittede
dager_utvidet=array(range(100))
y_utvidet = tilpasning(dager_utvidet)

#plotter utvidet versjon av den tilpassede kurven
plot(dager_utvidet,y_utvidet,label="Tilpasset kurve")
axhline(2200000,color="black",linestyle="--")
title("Utvidet regresjonskurve")
xlabel("Dager")
ylabel("Smittede individer")
grid()
legend()
show()
```

### 4. Drøfting

#### Modell 1:
I modell 1 øker antall smittede relativt sakte i starten, men etter omtrent 3 dager ser vi at det øker raskere. Grafen øker helt til dag 5, der den har et toppunkt på omtrent 90 000 mennesker, i en populasjon på 124 000. Etter dette begynner antallet smittede å synke. I denne modellen antar vi at man blir imun etter man har hatt sykdommen, og vi ser at antallet personer som har blitt friske igjen passerer antallet mottakelige omtrent på dag 5, da smitten begynner å snu. På grunn av det minkende antallet mottakelige har sykdommen vanskeligere for å spre seg blant befolkningen. 

Modell 1 er relativ enkel, og vil ikke kunne brukes til å spå smittespredning spesielt nøyaktig, men kan raskt gi en relativt godt estimat. I en såpass enkel modell som denne har vi gjort en del antakelser. N er konstant gjennom hele perioden, som betyr at vi ignorerer fødsler og innflyttinger. Det betyr også at vi antar at ingen dør av sykdommen. Modellen tar heller ikke hensyn til at en befolkning består av individer som lever forskjellige liv, og parameterene som b baserer seg på gjennomsnittlig smitte. Vi tar for eksempel ikke hensyn til at noen deler av befolkninger bor i tettere bebodde områder enn andre. 

Modellen kan tilpasses forskjellige smittesituasjoner ved å justere parameterene b og f. Dersom vi øker b vil antallet personer som blir smittet ved hvert tidssteg øke, men vi vil også nå flokkimmunitet raskere. Om vi øker f vil de syke bli friske igjen raskere, som vil gjøre at det vil være færre syke til enhver tid, og smitten vil spre seg raskere. Ved å velge riktige verdier for disse kan vi tilpasse modellen vår til aktuelle smittesituasjoner, for å forutse en mulig fremtidig utvikling. Denne modellen har passer kanskje ikke aller best til langsiktig smittemodellering, fordi den ikke tar hensyn til at befolknignen kan tilpasse seg situasjonen (bedre hygiene, holde avstand, munnbind osv...), men kan fungere greit over kortere perioder som denne. 

#### Modell 2:
Modell 2 er i likhet med modell 1 en SIR-smittemodell, og de er i stor grad like. Forskjellen her er at b og f ikke er konstanter, men funksjoner av tiden. I praksis gjør dette at både smittetallet og bedringsraten forandres underveis i smitteutviklingen. For f kan dette være en medisin. I dette tilfellet ser vi at kurven for de friskemeldte øker raskere etter dag 4, som er da medisinen begynner å virke. Medisinen har en litt varierende effekt på forskjellige mennesker, som gjørt at grafen svinger litt mer. Dette reflekteres også i antall smittede, som vi ser øker når effekten av medisinen, og derfor også f, er på sitt laveste. 

Sykdommen spres også på en annen måte nå, på grunn av at smittetallet er en funkskjon av t. Den spres nå raskere i starten, men smittetallet men smittetallet begynner å minke etterhvert, som gjør at de smittede etter en stund vil smitte færre av de som er rundt dem. Toppen her nås alleredde etter 2-3 dager, og er på nesten 11 000 smittede. Men vi ser også at antallet smittede etter 10 dager er mye lavere i modell 2.  

Denne modellen har fordelen at vi kan tilpasse parameterene slik at de blir likere vår situasjon, som gjør det mulig med litt mer nøyaktige verdier. For eksempel kan man tilpasse b slik at den blir mindre etterhvert på grunn av smittevernstiltak, som ville vært relevant til koronapandemien. Å bruke et delt funksjonsuttrykk som i f åpner også for flere tilpassninger. På grunn at f og b ikke er konstanter kan modell 2 passe bedre til å modellere smittesituasjoner over lengere tid, da du kan tilpasse verdiene av parameterene med tiden. Men modellen gir forstatt mer et røft estimat heller enn nøyaktige tall, og du kan lage bedre modeller der du blant annet kan basere smittetallet på flere individuelle variabler. I likhet med modell 1 antar vi også her at befolkningen holder seg konstant, og at sykdommen ikke er dødelig. Hverken modell 1 eller 2 har med vaksinering heller, som kan gjøre at folk ikke lenger er mottakelige, uten at de gjennomgår en smittsom fase.
